```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
housing['income_cat']=pd.cut(housing['median_income'], bins=[0., 1.5,3.0, 4.5, 6.0, np.inf], labels=[1,2,3,4,5])
housing.head()
```

```python
split=StratifiedShuffleSplit(n_splits=1, random_state=42, test_size=.2)
```

```python
# this a generator and needs to be looped over or list(split.split(housing,housing['income_cat']))[0][0].tolist()
for train_index, test_index in split.split(housing,housing['income_cat']):
    strat_train_set=housing.iloc[train_index]
    strat_test_set=housing.iloc[test_index]
```

```python
for set_ in (strat_train_set,strat_test_set):
    set_.drop(["income_cat"],axis=1, inplace=True)
```

# Pipeline

```python
from sklearn.base import BaseEstimator, TransformerMixin

```

```python
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

#class CombineAttributesAdder(BaseEstimator, TransformerMixin): #TMI is only for fit_transform, not required
class CombineAttributesAdder(BaseEstimator):
    def __init__(self, add_bedrooms_per_room=True):
        self.add_bedrooms_per_room=add_bedrooms_per_room
    def fit(self, X, y=None):
        return self #Nothing to do here. Estimator is not fitted
    def transform(self, X, y=None):
        rooms_per_household = X[:,rooms_ix]/X[:,households_ix]
        population_per_household=X[:,population_ix]/X[:,households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:,bedrooms_ix]/X[:,rooms_ix]  
            return np.c_[X,rooms_per_household,population_per_household,bedrooms_per_room]
        else:
            return np.c_[X,rooms_per_household,population_per_household]
```

```python
housing.columns
```

```python
attr_adder=CombineAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
```
